```python
import pandas as pd 
import mlflow 
import mlflow.sklearn
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.datasets import fetch_california_housing
```

```python
from urllib.parse import urlparse
```

```python
housing = fetch_california_housing()
```

```python
housing
```

```python
data = pd.DataFrame(housing.data,columns= housing.feature_names)
data
```

```python
data['price'] = housing.target
```

```python
data
```

```python
## Depandant + indepandant values 
```

```python
X = data.drop(columns = ["price"])
```

```python
X
```

```python
y = data['price']
```

```python
y
```

###### Train test split

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.20)
```

##### Hyperparameter_tuning 

```python
def hyperparametr_tuning(x_train,y_train,param):
    rf = RandomForestRegressor()
    grid_search = GridSearchCV(estimator=rf,param_grid=param,cv=3,n_jobs=-1,verbose=2,scoring="neg_mean_squared_error")
    grid_search.fit(x_train,y_train)

    return grid_search      
```

##### Model schemas  define with signatures 

```python
from mlflow.models import infer_signature
```

```python
signature = infer_signature(x_train,y_train)
```

```python
# define  prameters 
```

```python
param = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
```

```python
## mlflow exp
```

```python
with mlflow.start_run():
    #hypermarater tuning 
    grid_search = hyperparametr_tuning(x_train,y_train,param)
    # get best model 
    best_model = grid_search.best_estimator_
    ## evalute the model 

    y_pred = best_model.predict(x_test)
    mse = mean_squared_error(y_test,y_pred)
```

```python
# Log all best parameters in one line
mlflow.log_params(grid_search.best_params_)

# Log your metric
mlflow.log_metric("mse", mse)
```

```python
## tracking ui 
```

```python
mlflow.set_tracking_uri(uri = "http://127.0.0.1:5000")
tracking_url_type_store = urlparse(mlflow.get_tracking_uri()).scheme

if tracking_url_type_store != "file":
    mlflow.sklearn.log_model(best_model,"model",registered_model_name= "best randomforestmodel")

else :
    mlflow.sklearn.log_model(best_model,"model",signature=signature)


print (f"best hyperparameter :{grid_search.best_params_}")
print (f"mean sqr error :{mse}")
```
